The following is a code to explore a simple type Ia supernova model 

```python
import sys
!{sys.executable} -m pip install --quiet wnutils

import wnutils.xml as wx 
import numpy as np
import matplotlib.pyplot as plt

xml = wx.Xml('out_Ia_weak_1_59e9.xml')
```

## Plot Temperature vs. time 

The following shows the temperature of the single zone vs. time. The model begins with a temperature of 1.4 billion Kelvin. The reignition of nuclear fuel begins the runaway nuclear fusion that results in the type Ia supernova - seen in the plot by the temperature spike. Notable features of the plot are the peak at which the material reaches quasi-statistical equilibrium before using energy to disintegrate $^56$Ni to reach full nuclear statistical equilibrium. 

```python
xml.plot_property_vs_property('time','t9',xscale = 'log', xlim = [1e0,1e1], xlabel = 'Time (s)', ylabel = 'T/$10^9K$')
```

## Plot Entorpy per nucleon vs time


```python
xml.plot_property_vs_property('time','entropy per nucleon', xscale = 'log', xlim = [1e0,1e2], ylim=[0,10],xlabel = 'Time (s)', 
                              ylabel = 'Entropy/Nucleon')
```

## Plot chemical potential


```python
xml.plot_property_vs_property('time','muekT',xlabel='Time (s)',ylabel = '$\mu_e\'/k_BT$',
                              xlim = [1e-6,1e2],xscale = 'log')
```
